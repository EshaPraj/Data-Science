```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#seaborn - ML VISUALIZATION LIBRARY
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
#matplotlib : barplot, boxplot, scatter plot, histogram, ....

#seaborn : scatterplot, pairplot, countplot, violinplot, swarmplot, heatmap, boxplot....
```

```python
#sns.scatterplot()
#sns.pairplot()
#sns.countplot()
#sns.violinplot()
#sns.swarmplot()
#sns.heatmap()
#sns.boxplot()
sns.scatterplot(x = "total_bill", y = "tip", data = tips);
```

```python
from scipy.stats import pearsonr
pearsonr(tips.total_bill, tips.tip)
```

```python
sns.pairplot(tips);
```

```python
#heatmap
corr = tips.corr(numeric_only = True)
sns.heatmap(corr, annot = True, cmap = "coolwarm");
```

```python
#boxplot
sns.boxplot(x = "day", y = "total_bill", data = tips, palette = "pastel");
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips, palette = "viridis");
```
